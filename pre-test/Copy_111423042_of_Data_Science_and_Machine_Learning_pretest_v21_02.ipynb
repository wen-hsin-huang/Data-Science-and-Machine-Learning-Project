1.lowercase : 將英文字母改為小寫字母。

為什麼做lwercase或不做lowercase?

在大部分狀況，會在前處理時進行lowercase的動作，可幫助後續進行NLP處理上的效能[1]，也因此大部分的人都會將自己的所有資料都進行lowercase的處理。[2]lowercase經常被忽略，但是最基本最簡單的前處理方式，可以協助資料探勘與進行NLP處理時，讓輸出的結果有一致性。當在進行neural network等模型訓練時，如果訓練集過小，則會造成在進行test時因為大小寫差異導致無輸出結果或是輸出較少的情形，舉例TAIWAN、Taiwan、taiwan、TAiwan，可能被視為不同的內容。[3]

不做lowercase的原因，在辨識JAVA與python的程式語言，其撰寫規則不同，JAVA的格式與python的格式，可以使用大小寫來進行初步判斷，如此時已進行lowercase處理，則少了一項特徵。[3]

2.decontracting

因為英文的語法，有縮寫與一般寫法的不同，為了讓兩種文字為同一種意思，會將其進行decontracting，以避免在進行text classification task時，同樣意思卻視為不同語詞，舉例如"can't"等同於"can not"，"yor're"等同於"you are"。[4]


3.remove tags, punctuations, numbers

當在進行資料檢索、text classification時我們會刪去tags, punctuations, numbers，因為在進行分類時，非文字以外的數字不視為特徵，舉例在進行google搜尋時，如果搜尋"10種機器學習的方式"，則可能會出現多種的機器學習方式頁面，不限定於只有10種，可能會大於等於或小於，而下方則會顯示關鍵字缺少:10，利用，利用remove tags, punctuations, numbers，可以上，可以讓資料檢索、text classification的範圍不會被文字以外的特徵限縮。[5]

4.tokenization

是將字串進行切割，當我們要判斷的text不是字母、句子、文章而是一個單字時，則使用此方法進行切割，以不同的語言則會有不同的方式進行，舉例I love text classificatioon.是一個句子，利用tokenizaation則可將句子切割成"I"、"love"、"text"、"classification"，以進行後續NLP處理[6]

5.stopword removal

當我們利用google進行資料檢索時，通常我們會直接下關鍵字進行搜尋，或是整個句子，但其實像是冠詞、動詞等字並非重要的檢索資訊，則可以進行前處理刪除，已找到更多的資訊，舉例: what is stopword removal，我們可以簡化成 stopword removal以進行後續搜尋。通常可以應用於search systems、text classification。利用stopword removal的動作可以減少搜尋不相關的關鍵字，以提高搜尋或是分類速度。[3]

6.lemmatization

此項前處理和stemming相似，差別在於進行的是如"good"、"well"是一樣的意思，將其整合視為一種字詞，通常會使用dictionary，如: WordNet，以及一些規則進行處理，但其處理成效較stemming小，並且在進行neural network時，此處理會花費比stemming更多的時間，但影響成效不大，是否須對text進行lemmatization，可取決於使用的演算法以及時間成本是否有益。[3]

7.stemming

字詞的不同語態整合，例如：classes，class，一個為複數單位，一個為單一單位，但其實是指一樣的字詞，當我們在搜尋 "deep learning classes"，我們希望也能找到"deep learn class"等相關字詞，因此需要使用stemming來整合，以進行後續NLP處理，而進行stemming的格式有很多種，其中最常見的是Porters Algorithm[3]

## 3-2. Define a converting function to compute tf-idf vector from a list of ducoments. (40%)

```
documents = ['terrible service this time','terrible terrible service','most terrible service','terrible service and experience','what a terrible service','so terrible service experience','what a terrible disappointment','what a terrible place','this time it was so horrible','the staff was horrible']
```

```
'''
Answer here
you can define other functions to support the defined function if you need.

TF-IDF dataframe show as the following table.
'''
import numpy as np
import math
def get_tf(doc):
    tf_dictionary = {}
    for word in doc:
        if word in tf_dictionary:
            tf_dictionary[word] += 1
        else:
            tf_dictionary[word] = 1
    for word in tf_dictionary:
        tf_dictionary[word] = tf_dictionary[word] / float(len(doc))
    return tf_dictionary

def get_idf(documents):
  idf_dict = {}
  word_lists = [doc.split() for doc in documents]
  unique_words = set(word for word_list in word_lists for word in word_list)
  words = np.concatenate(word_lists)
  unique_words = np.unique(words)
  for the_term_index in range(len(unique_words)):
    term = unique_words[the_term_index]
    n = sum(1 for doc in documents if term in doc.split()) ####doc.split()非常重要，將會變成找字母或是找單字
    N = len(documents)
    idf_dict[term] = math.log(N / n)  
  return idf_dict

def computTFIDF(documents):
  tfidf = []
  idf = get_idf(documents)
  for doc in documents:
      tf = get_tf(doc.split())
      tfidf_doc = {'terrible':float(0), 'service':float(0), 'this':float(0),'time':float(0), 'most':float(0),'and':float(0),'experience':float(0),'what':float(0),'a':float(0)
      ,'so':float(0),'disappointment':float(0),'place':float(0),'it':float(0),'was':float(0),'horrible':float(0),'the':float(0),'staff':float(0)}
      for term, val in tf.items():
          tfidf_doc[term] = val * idf[term]
      tfidf.append(tfidf_doc)

  return np.array(tfidf).tolist()
```

## 3-3. Define a scoring function to compute the cosine similarity between two input vector. (30%)

```
'''
Answer here
return the cosine similarity between the given two vectors
Apply the function which you designed to all sentences, and show your scoring results as the following table.
'''
def cosine_sim(vec_a, vec_b):
  #ref: https://stackoverflow.com/questions/39608421/showing-valueerror-shapes-1-3-and-1-3-not-aligned-3-dim-1-1-dim-0
  dot_product = float(np.dot(vec_a, vec_b.T))
  mag_a = float(math.sqrt(np.dot(vec_a, vec_a.T)))
  mag_b = float(math.sqrt(np.dot(vec_b, vec_b.T)))
  score = float(dot_product / (mag_a * mag_b))
  return score
```
